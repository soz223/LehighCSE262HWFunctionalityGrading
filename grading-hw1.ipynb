```python
import os
import subprocess

# Path to the base directory where all the student folders are located
base_dir = os.path.abspath("./repo/homework-1")  # Using absolute path to avoid relative path issues

result_dir = os.path.join(base_dir, "results.txt")

if not os.path.exists(result_dir):
    with open(result_dir, "w") as f:
        f.write("")  # Create the file if it doesn't exist

# Function to run cargo test and check results
def run_cargo_test_in_folder(folder_path):
    try:
        # Change to the homework-1 directory
        os.chdir(folder_path)

        # Run the `cargo test` command and capture the output
        result = subprocess.run(['cargo', 'test'], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

        # Check if cargo test executed successfully
        if result.returncode == 0:
            output = result.stdout

            # Find the line containing the test result summary
            for line in output.splitlines():
                if "test result:" in line:
                    # Parse the result to extract passed and total tests
                    passed_tests = int(line.split("passed;")[0].split()[-1])
                    failed_tests = int(line.split("failed;")[0].split()[-1])
                    
                    total_tests = passed_tests + failed_tests
                    
                    print(f"{folder_path}: {passed_tests}/{total_tests} tests passed.")
                    with open(result_dir, "a") as f:
                        f.write(f"{folder_path}: {passed_tests}/{total_tests} tests passed.\n")
                    return True
            
            print(f"{folder_path}: No test results found.")
        else:
            print(f"{folder_path}: cargo test failed to execute.")
    except Exception as e:
        print(f"Error in {folder_path}: {e}")
    finally:
        # Return to the original directory
        os.chdir(base_dir)

# Loop through each student's directory and run tests
def main():
    # Get the list of student directories
    student_folders = peoples

    # write something here to record if all people are tested or not
    is_test_array = [False] * len(student_folders)
    

    # Iterate through each student folder
    for student_folder in student_folders:
        # Construct the path to the homework-1 directory, normalize it for cross-platform compatibility
        hw1_folder = os.path.normpath(os.path.join(base_dir, student_folder + '-cse262', 'homework-1'))

        # Check if homework-1 directory exists
        if os.path.exists(hw1_folder):
            is_test_successful = run_cargo_test_in_folder(hw1_folder)
            # if the cargo test is successful, set the is_test_array to True
            if is_test_successful:
                is_test_array[student_folders.index(student_folder)] = True
        else:
            print(f"{student_folder} does not have a homework-1 folder.")

if __name__ == "__main__":
    main()
```
